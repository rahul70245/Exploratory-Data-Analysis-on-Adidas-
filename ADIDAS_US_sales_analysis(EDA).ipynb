 **UNIVARIATE ANALYSIS**







```python
# numerical columns
```

```python
#minimum value of all numerical columns

df.agg({'Price per Unit':'min','Units Sold':'min','Total Sales':'min','Operating Profit':'min'})
```

```python
#maximum value of all numerical columns

df.agg({'Price per Unit':'max','Units Sold':'max','Total Sales':'max','Operating Profit':'max'})
```

```python
#average value of all numerical columns

df.agg({'Price per Unit':'mean','Units Sold':'mean','Total Sales':'mean','Operating Profit':'mean'})
```

```python


# distribution, skewness, mean, median of numerical features

n=['Price per Unit','Units Sold','Total Sales','Operating Profit']

for col in n:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df[col]
    feature.hist(bins=50, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
# for distribution we can also use distplot

sns.distplot(df['Price per Unit'])
```

```python
#categorical columns
```

```python
#contribution of retailer in sales

[df['Retailer'].value_counts()]
```

```python
total=df['Retailer'].value_counts().sum()
def x(n):
  return (n/total)*100
```

```python
#in percent retailer contribution in sales

round(df['Retailer'].value_counts().apply(x)).astype(str)+'%'
```

```python
#pie graph showing the same

round(df['Retailer'].value_counts().apply(x)).plot(kind='pie')
```

```python
#region wise contribution in sales
round(df['Region'].value_counts().apply(x)).astype(str)+'%'
```

```python
#pie graph showing same

round(df['Region'].value_counts().apply(x)).plot(kind='pie')
```

```python
#product wise contribution in sales

round(df['Product'].value_counts().apply(x)).astype(str)+'%'
```

```python
#sales method wise contribution in sales

round(df['Sales Method'].value_counts().apply(x)).astype(str)+'%'
```

```python
#pie graph showing the same

round(df['Sales Method'].value_counts().apply(x)).plot(kind='pie')
```

**BIVARIATE ANALYSIS**

```python
#state wise sales

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['State'])['Total Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('State Wise Sales')
plt.xlabel('State')
plt.ylabel('Sales')
```

```python
#region wise sales

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Region'])['Total Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Region Wise Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
```

```python
#product wise sales

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Product'])['Total Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Product Wise Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
```

```python
#sales method wise sales

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Sales Method'])['Total Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Sales Method Wise Sales')
plt.xlabel('Mode')
plt.ylabel('Sales')
```

```python
#retailer wise sales

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Retailer'])['Total Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Retailer Wise Sales')
plt.xlabel('Retailer Name')
plt.ylabel('Sales')
```

```python
#highest sales product name and retailer name

df.loc[df['Total Sales'].idxmax(),['Product','Retailer']]
```

```python
#lowest sales product name and retailer name

df.loc[df['Total Sales'].idxmin(),['Product','Retailer']]
```

```python
#state wise profit

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['State'])['Operating Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('State Wise Profit')
plt.xlabel('State')
plt.ylabel('Profit')
```

```python
#region wise profit

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Region'])['Operating Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Region Wise Profit')
plt.xlabel('Region')
plt.ylabel('Profit')
```

```python
#product wise profit

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Product'])['Operating Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Product Wise Profit')
plt.xlabel('Product')
plt.ylabel('Profit')
```

```python
#sales method wise profit

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Sales Method'])['Operating Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Sales Method Wise Profit')
plt.xlabel('Mode')
plt.ylabel('Profit')
```

```python
#retailer wise profit

plt.rcParams['figure.figsize']=(20,10)
df.groupby(['Retailer'])['Operating Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Retailer Wise Profit')
plt.xlabel('Retailer Name')
plt.ylabel('Profit')
```

```python
#highest profit product name and retailer name

df.loc[df['Operating Profit'].idxmax(),['Product','Retailer']]
```

```python
#lowest profit product name and retailer name

df.loc[df['Operating Profit'].idxmin(),['Product','Retailer']]
```

```python
#avg profit for different product type

df.groupby(['Product'])['Operating Profit'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
# avg profit distribution for different product type in january 2020 and 2021

df['month']=df['Invoice Date'].dt.month
df1=df[df['month']==1]


df1.groupby(['Invoice Date','Product'])['Operating Profit'].mean().unstack().plot(kind='box')
```

```python
#analysis of profit and sales by product


plt.rcParams['figure.figsize']=(20,8)
df.groupby(['Product'])['Operating Profit','Total Sales'].sum().plot()
```
